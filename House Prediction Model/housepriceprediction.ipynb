```python
import pandas as pd
house = pd.read_csv(r"D:\User Do Not Del\Desktop\Takshat\ML\House Prediction\house_prices.csv")
house
```

```python
import numpy as np
house = house.drop(columns=["Title","Description","Status","Floor","Society","Ownership","Dimensions","Plot Area"])
house
```

```python
house.isnull().sum()
```

```python
columns_to_clean = ['Price (in rupees)','Transaction','Furnishing','facing','overlooking','Bathroom','Balcony','Car Parking']  # Specify columns you want to clean
house[columns_to_clean] = house[columns_to_clean].fillna(value=0)
house
```

```python
house['location'] = house['location'].astype('category')
house['location'] = house['location'].cat.codes

house['Transaction'] = house['Transaction'].astype('category')
house['Transaction'] = house['Transaction'].cat.codes

house['Furnishing'] = house['Furnishing'].astype('category')
house['Furnishing'] = house['Furnishing'].cat.codes

house['facing'] = house['facing'].astype('category')
house['facing'] = house['facing'].cat.codes

house['overlooking'] = house['overlooking'].astype('category')
house['overlooking'] = house['overlooking'].cat.codes

house['Car Parking'] = house['Car Parking'].astype('category')
house['Car Parking'] = house['Car Parking'].cat.codes

house
```

```python
def convert_amount(value):
    if isinstance(value, str):
        if 'lac' in value.lower():
            return float(value.lower().replace(' lac', '').strip()) * 100000
        elif 'cr' in value.lower():
            return float(value.lower().replace(' cr', '').strip()) * 10000000
        else:
            return 0  # or return value as is, or any other placeholder for invalid values
    else:
        return value

# Apply the function to the 'amount' column
house['Amount(in rupees)'] = house['Amount(in rupees)'].apply(convert_amount)
house
```

```python
X = house.drop(columns=['Price (in rupees)'])
X
```

```python
y = house['Price (in rupees)']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.5,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred_train = lr.predict(X_train)
y_pred_train
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_train,y_pred_train)
plt.xlabel('Actual profit')
plt.ylabel('Predicted Profit')
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y_train, y_pred_train)
```

```python
y_pred_test = lr.predict(X_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred_test)
plt.xlabel('Actual profit')
plt.ylabel('Predicted Profit')
plt.show()
```

```python
r2_score(y_test, y_pred_test)
```
